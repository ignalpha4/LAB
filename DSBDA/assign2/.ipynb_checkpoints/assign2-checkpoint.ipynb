```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

col = ['math score', 'reading score' , 'writing score','placement score']
fig = plt.figure(figsize =(10, 7))
df.boxplot(col)
plt.show()
```

```python
#3.2 handling of outliers
```

```python
df_stud=df

x19thpercentile=np.percentile(df_stud['math score'],90)
b=np.where(df_stud['math score']>x19thpercentile,x19thpercentile,df_stud['math score'])
```

```python
print("new array:",b)
```

```python
df_stud.insert(1,'m score',b,True)
df_stud
```
