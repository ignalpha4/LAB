```python
#4.3 Detecting outliers using Z-Score:
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(df['math score']))

print(z)
```

```python
threshold = 0.18

sample_outliers = np.where(z <threshold)
sample_outliers
```

```python
#4.4 Detecting outliers using Inter Quantile Range(IQR):

import numpy as np

sorted_rscore= sorted(df['reading score'])

sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1   #for calculating value of IQR
 
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
#printing outliers

r_outliers = []
for i in sorted_rscore:
    if (i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
print(r_outliers)
```

```python
#3.2 handling of outliers
```

```python
# 1.Trimming/removing the outlier:
```

```python

new_df=df
for i in sample_outliers:
    new_df.drop(i,inplace=True)
new_df
```

```python
# 2.  Quantile based flooring and capping:

df=pd.read_csv("demo1.csv")


df_stud=df

nintteenth_percentile=np.percentile(df_stud['math score'],90)
b=np.where(df_stud['math score']>nintteenth_percentile,nintteenth_percentile,df_stud['math score'])
```

```python
median=np.median(sorted_rscore)  #Calculate the median of reading score by using sorted_rscore
median 
```

```python
# Replace the upper bound outliers using median value

refined_df=df

refined_df['reading score'] = np.where(refined_df['reading score'] >upr_bound, median,refined_df['reading score'])

refined_df
```

```python
#Replace the lower bound outliers using median value

refined_df['reading score'] = np.where(refined_df['reading score'] <lwr_bound, median,refined_df['reading score'])
refined_df
```
